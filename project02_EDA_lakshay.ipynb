```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Project Task-01: Print the shape of the dataset
intake_df = pd.read_csv('DataSet_Animal_Center_Intakes.csv')
print("Shape of the intake dataset:", intake_df.shape)
outcome_df = pd.read_csv('DataSet_Animal_Center_Outcomes.csv')
print("Shape of the outcome dataset:", outcome_df.shape)
```

```python
# Project Task-02: Which column has null values?
null_columns = intake_df.columns[intake_df.isnull().any()]
print("Columns with null values:", null_columns)
```

```python

# Project Task-03: Remove duplicate rows from the intake dataset
intake_df.drop_duplicates(inplace=True)
```

```python

# Project Task-04: Rename the column names
column_rename = {
    'Animal ID': 'id',
    'DateTime': 'date_time',
    'MonthYear': 'month_year',
    'Found Location': 'found_location',
    'Intake Type': 'intake_type',
    'Intake Condition': 'intake_condition',
    'Animal Type': 'animal_type',
    'Sex upon Intake': 'sex_upon_intake',
    'Age upon Intake': 'age_upon_intake',
    'Breed': 'breed'
}
intake_df.rename(columns=column_rename, inplace=True)
```

```python

# Project Task-05: Remove MonthYear column
intake_df.drop(columns='month_year', inplace=True)
```

```python
# Project Task-06: Extract month and year from the Datetime column
intake_df['month'] = pd.to_datetime(intake_df['date_time']).dt.month
intake_df['year'] = pd.to_datetime(intake_df['date_time']).dt.year
```

```python
# Project Task-07: Pie chart for year-wise animal intake from 2014 to 2021
year_wise_count = intake_df['year'].value_counts()
plt.pie(year_wise_count, labels=year_wise_count.index, autopct='%1.1f%%')
plt.title('Year-wise Animal Intake')
plt.show()
```

```python
# Project Task-08: Bar chart for month-wise animal intake
month_wise_count = intake_df['month'].value_counts()
plt.bar(month_wise_count.index, month_wise_count.values)
plt.xlabel('Month')
plt.ylabel('Animal Intake Count')
plt.title('Month-wise Animal Intake')
plt.show()
```

```python
# Project Task-09: Bar chart for animal intake methods wise counting
intake_methods_count = intake_df['intake_type'].value_counts()
plt.bar(intake_methods_count.index, intake_methods_count.values)
plt.xlabel('Intake Method')
plt.ylabel('Count')
plt.title('Animal Intake Methods Count')
plt.show()
```

```python

# Project Task-10: Conditions of animals taken into the shelter
animal_conditions = intake_df['intake_condition'].unique()
print("Conditions of animals in the shelter:", animal_conditions)
```

```python
# Project Task-11: Types of animals taken into the shelter and their total count
animal_types_count = intake_df['animal_type'].value_counts()
print("Types of animals and their counts:")
print(animal_types_count)
```

```python
# Project Task-12: Top 10 breeds in the animal shelter (bar plot)
top_10_breeds = intake_df['breed'].value_counts().head(10)
plt.bar(top_10_breeds.index, top_10_breeds.values)
plt.xlabel('Breed')
plt.ylabel('Count')
plt.title('Top 10 Breeds in Animal Shelter')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 13: Top 20 common names of animals in the animal shelter (bar plot)
print(intake_df.columns)  # Display all column names to verify the correct column name

# Assuming the correct column name is 'Animal_Type'
top_20_names = intake_df['Name'].value_counts().head(20)
plt.bar(top_20_names.index, top_20_names.values)
plt.xlabel('Name')
plt.ylabel('Count')
plt.title('Top 20 Common Names of Animals')
plt.xticks(rotation=90)
plt.show()


```

```python
# Project Task-14: Bar chart for "Outcome Type" wise count of animals using the Animal_Center_Outcomes.csv dataset
outcome_df = pd.read_csv('DataSet_Animal_Center_Outcomes.csv')
outcome_type_count = outcome_df['Outcome Type'].value_counts()
plt.bar(outcome_type_count.index, outcome_type_count.values)
plt.xlabel('Outcome Type')
plt.ylabel('Count')
plt.title('Outcome Type Wise Count of Animals')
plt.xticks(rotation=90)
plt.show()
```
